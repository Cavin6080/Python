# Importing the libraries and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("SampleSuperstore.csv")
dataset.head()
```

```python
dataset =dataset.drop(labels=["Postal Code"], axis=1)
dataset.head(20)
```

```python
#Checking types
dataset.dtypes
```

```python
#Gives allstatisical details 
dataset.describe()
```

```python
#Checking the number of rows and columns
dataset.shape
```

```python
#Checking if any missing data found in the dataset
missing_data = dataset.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

```python
dataset.corr()
```

```python
#Checking duplicates
dataset.duplicated().sum()
```

```python
#Droping duplicates
dataset.drop_duplicates(keep = 'first', inplace = True)
```

```python
dataset.duplicated().sum()
```

## Visualization of data with pairplot

```python
import seaborn as sns
sns.pairplot(dataset)
```

## Visualizing the Ship mode

```python
plt.axis("equal")
plt.pie(dataset['Ship Mode'].value_counts(), labels = dataset['Ship Mode'].value_counts().index, radius = 1.5, autopct = '%0.2f%%',
       shadow = True, explode = [0,0.2,0.1,0.09], startangle = 45)
plt.show()
```

## Visualizing the Ship mode in terms of profit 

```python
grp1 = dataset.groupby(['Ship Mode'], as_index = False)['Profit'].sum()
sns.barplot(grp1['Ship Mode'], grp1["Profit"])
```

### Standard class makes most profit

## Visualizing the Segment

```python
plt.axis("equal")
plt.pie(dataset['Segment'].value_counts(), labels = dataset['Segment'].value_counts().index, radius = 1.5, autopct = '%0.2f%%',
       shadow = True,explode = [0,0.2,0.1],  startangle = 45)
plt.show()
```

## Visualizing the Segment in terms of profit 

```python
grp2 = dataset.groupby(['Segment'], as_index = False)['Profit'].sum()
sns.barplot(grp2['Segment'], grp2["Profit"])
```

### Consumer segment makes the most profit here

## Commodities Sales analysis

```python
newdata = dataset.groupby("Category")["Quantity"].sum()
plt.axis("equal")
plt.pie(newdata.values, labels = newdata.index, radius = 1.5, autopct = '%0.2f%%',
       shadow = True,  startangle = 0)
plt.show()
```

## commodities prices analysis

```python
commodities=dataset.groupby("Sub-Category").Sales.agg(["mean","max","min","std"])
commodities
```

## Top Regions in terms of Sales and Profit

```python
region_s = dataset.groupby("Region").Sales.sum().nlargest(n=200)
region_p = dataset.groupby("Region").Profit.sum().nlargest(n=200)
region_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, sort_columns=True,color='red')
region_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green',sort_columns=True)
plt.xlabel('Region',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Regions in terms of Sales and Profit",fontsize = 15)
plt.show()
```

## Top  Categories  in terms of Sales and Profit

```python
cat_s = dataset.groupby("Category").Sales.sum().nlargest(n=200)
cat_p = dataset.groupby("Category").Profit.sum().nlargest(n=200)
cat_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, sort_columns=True,color='red')
cat_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green',sort_columns=True)
plt.xlabel('Categories',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Categories in terms of Sales and Profit",fontsize = 15)
plt.show()
```

## Top Sub-Categories in terms of Sales and Profit

```python
#Sub-Category
scat_s = dataset.groupby("Sub-Category").Sales.sum().nlargest(n=200)
scat_p = dataset.groupby("Sub-Category").Profit.sum().nlargest(n=200)
scat_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, sort_columns=True,color='red')
scat_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green',sort_columns=True)
plt.xlabel('Sub-Categories',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Categories in terms of Sales and Profit",fontsize = 15)
plt.show()
```

# Top states in terms of Sales and Profit

```python
state_s = dataset.groupby("State").Sales.sum().nlargest(n=200)
state_p = dataset.groupby("State").Profit.sum().nlargest(n=200)
state_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, sort_columns=True,color='red')
state_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green',sort_columns=True)
plt.xlabel('State',fontsize = 10)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top states in terms of Sales and Profit",fontsize = 15)
plt.show()
```

# Top cities in terms of Sales and Profit

```python
city_s = dataset.groupby("City").Sales.sum().nlargest(n=50)
city_p = dataset.groupby("City").Profit.sum().nlargest(n=50)
city_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, sort_columns=True,color='red')
city_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green',sort_columns=True)
plt.xlabel('Cities',fontsize = 10)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Cities in terms of Sales and Profit",fontsize = 15)
plt.show()
```
