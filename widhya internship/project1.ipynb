```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv")
df.head()
```

```python
df1 = df.drop(['State/UnionTerritory'], axis = 1)
```

```python
df1.head()
df1.shape
```

```python
sumcol = ['ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths' ]
df1['sum'] = df1[sumcol].sum(axis = 1, skipna = False)
df1.head()
```

```python
df1 = df1.groupby('Date', sort = False)
df1.head()
```
